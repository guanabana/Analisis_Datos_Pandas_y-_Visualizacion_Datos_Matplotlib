```python
#cuantos directores distintos - Agarrar columna y borrar duplicados = VALORES ÚNICOS
valores_unicos = df['director'].drop_duplicates()
print(valores_unicos)
```

```python
#AGRUPAR por director y contar tirulos -> De todos los directores distintos cuantas pelis hicieron 
#Agrupor por (columna que quiero)[la columna por la que cuento ] luego operacion
conteo_por_director = df.groupby('director')['title'].count()
print(conteo_por_director)
```

```python
#imprimirlo
for director, cantidad in conteo_por_director.items():
    print("Director:", director)
    print("Cantidad de títulos:", cantidad)
    print("-" * 30)
```

```python
# Agrupar por año y contar la cantidad de títulos en cada año
#cantidad de pelis que se hicieron por año
conteo_por_año = df.groupby('release_year')['title'].count()
conteo_por_año
```

```python
# Graficar el resultado usando PRIMITIVAS DE PANDA
plt.figure(figsize=(10, 6)) #TAMAÑO DE GRÁFICO

'''#LE DIGO TIPO DE GRAFICO BARRA Y LA SERIE POR DEFECTO A TOMAR EJEX .index "etiquetas" eje Y .values los valroes '''
plt.bar(conteo_por_año.index, conteo_por_año.values, color='green') 

plt.title('Cantidad de Títulos por Año', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Cantidad de Títulos', fontsize=14)

plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#<b>LLevamos el formato date_added a formato legible dd/mm/yyyy </b>
# CALCULAR TIEMPO ENTRE QUE SE REALIZO PELI release_year Y SE LANZÓ date_added
''' CONVERSION DE DATOS CON UNA COLUMNA NUEVA PARA UNIFICAR FORMATO STRING Y NUMERICO'''

df['date_added_new'] = df['date_added'].str.strip() #SACARLE ESPACIOS EN BLANCO

# Convertir la columna 'date_added' a tipo datetime especificando el formato correcto
df['date_added_new'] = pd.to_datetime(df['date_added_new'], format='%B %d, %Y')

#df['date_added_new'] = df['date_added_new'].dt.strftime('%Y%m%d')
df['date_added_new'] = df['date_added_new'].dt.strftime('%Y')
# Imprimir el DataFrame resultante
df.tail()
```

```python
'''AGRUPAR POR TIPO PELI Y POR TITULO'''
tipos_peliculas = df.groupby('type')['title'].count()
tipos_peliculas
```

```python
'''AGRUPAR POR PAIS --'''
tipos_peliculas = df.groupby('country')['title'].count()
tipos_peliculas
```

```python
'''MORAMLIZACIÓN DE DATOS PORQUE LOS DATOS VENIAN MAL EN FUENTE--- CUANTAS PELIS HAY POR PAIS'''

#CREAR COLUMNA NUEVA
df['country_new'] = df['country'].str.strip()

# "SPLITIAR"  POR COMA Y .EXPLODE = APLANAR EL ARRAY
df_countries = df.assign(country=df['country'].str.split(', ')).explode('country')
conteo_por_pais = df_countries['country'].value_counts().reset_index()
conteo_por_pais.columns = ['country', 'total_titles']
print(conteo_por_pais)
```

```python
'''QUIERO GRAFICAR EL NUEVO CAMPO'''
conteo_por_pais = conteo_por_pais.sort_values(by='total_titles', ascending=False)

# Tomar solo los primeros 10 países para graficar
top_10_paises = conteo_por_pais.head(10)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(top_10_paises['country'], top_10_paises['total_titles'], color='skyblue')
plt.title('Top 10 Países con Mayor Cantidad de Títulos', fontsize=16)
plt.xlabel('País', fontsize=14)
plt.ylabel('Cantidad de Títulos', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```
